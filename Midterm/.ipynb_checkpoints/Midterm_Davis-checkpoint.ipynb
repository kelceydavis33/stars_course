$P_0 =  k_{NR}[\frac{\rho_0}{\mu_R}]^{5/3}$

$\rho_0, m_0, r_0$ defined in problem

we consider that:

$m_{next} = \Delta m + m_{last}$

which gives

$\rho(r) = [\frac{P(r)}{k_{NR}}]^{3/5} \mu_{R}$

$m_{next} = 4 \pi r^2 \rho(r) \Delta r + m_{last}$

$P_{next} = \frac{-Gm(r) \rho(r)}{r^2} \Delta r + P_{last}$

```python
G =c.G.decompose(bases=u.cgs.bases).value
pi = np.pi
knr = 1e13
mur = 2
#In cm
R = 584000000
dr = 1e6
#Define initial values
r = 1e-12
m = 1e-12
rho = 6.19e6 
P = knr*((rho/mur)**(5/3))
#Creating radii for iteration
rvals = np.arange(r,R,dr)
#Define lists to store values
mvals = []
rhovals = []
Pvals = []
#mvals.append(m)
#rhovals.append(rhonew)
#Pvals.append(P)
#Loop through the radii values:
for rad in rvals:
    rhonew = ((P/knr)**(3/5))*mur
    m += 4*pi*(rad**2)*rhonew*dr
    P += ((-G*m*rhonew)/(rad**2))*dr
    #Add to lists
    mvals.append(m)
    rhovals.append(rhonew)
    Pvals.append(P)
```

Now let’s model the Sun. You’ll need all five stellar structure equations, and can assume the
following:

Calculate the density, enclosed mass, and temperature as a function of radius. Plot each versus

Calculate the density, enclosed mass, and temperature as a function of radius. 

We can start out again by looking at our equations:

**Our equations**:

mass distribution:

$\frac{dm}{dr} = 4 \pi r^2 \rho(r)$

hydrostatic equilibrium:

$\frac{dP}{dr} = \frac{-Gm(r) \rho(r)}{r^2}$

equation of state dominated by ideal gas:

$P = \frac{\rho}{\mu m_H}kT$

Energy transport:

$\frac{dT}{dr} = \frac{-3 \kappa \rho}{64 \pi \sigma T^3} \frac{L}{r^2}$

Energy generation by thermal fusion:

$\frac{dL}{dr} = 4 \pi r^2 \rho(r) [\epsilon_{nuc} - \epsilon{\nu}]$

but $\epsilon_{\nu} = 0 $ for proton proton chain domination




Changing out for deltas, we can form some basic equations:

$\Delta m = 4 \pi r^2 \rho(r) \Delta r$

$\Delta P = \frac{-Gm(r) \rho(r)}{r^2} \Delta r$

$\Delta T = \frac{-3 \kappa \rho}{64 \pi \sigma T^3} \frac{L}{r^2} \Delta r$

$\Delta L = 4 \pi r^2 \rho(r) [\epsilon_{nuc}] \Delta r$

$P = \frac{\rho}{\mu m_H}kT$



$\rho = \frac{kT}{P \mu m_H}$

and remembering that:

$\epsilon_{pp} = \epsilon_{0, pp} X^2 \rho T^4$

and 

$\mu = \frac{1}{2X + \frac{3}{4}Y + \frac{1}{2}Z}$

where X is the H fraction, Y is the He fraction, and Z is the fraction of heavy elements.

I grab the X, Y, and Z values from here:

https://en.wikipedia.org/wiki/Metallicity

```python
G =c.G.decompose(bases=u.cgs.bases).value
pi = np.pi
k = c.k_B.decompose(bases=u.cgs.bases).value
mh = c.u.decompose(bases=u.cgs.bases).value
X = 0.7381#0.7
Y = 0.2485#0.27
Z = 1-(X+Y)#0.134
mu = 1/((0.5*X) + (0.75*Y) + (0.5*Z) )
kappa = 0.20*(1+X)
sigma = c.sigma_sb.decompose(bases=u.cgs.bases).value
#In cm
R = c.R_sun.decompose(bases=u.cgs.bases).value
dr = 1e8
#Define initial values, making zero values just extremely small so that nothing weird happens with the math
r = 1e-12
m = 1e-12
L = 1e-12
T = 1.45e7
rho = 149
epsilon = 1.73e-5 * (X**2)*rho*(((1e-6)*T)**4)
P = (k*(T)*rho)/(mu*mh)
#Creating radii for iteration
rvals = np.arange(r,R,dr)
#Define lists to store values
mvals = []
rhovals = []
Pvals = []
Lvals = []
Tvals = []
#Loop through the radii values:
for rad in rvals:
    rhonew = (P*mu*mh)/(k*(T))
    epsilon = 1.73e-5 * (X**2)*rhonew*((1e-6*T)**4)
    m += 4*pi*(rad**2)*rhonew*dr
    P += ((-G*m*rhonew)/(rad**2))*dr
    L += 4*pi*(rad**2)*rhonew*epsilon*dr
    T += ((-3*kappa*rhonew)/(64*pi*sigma*((T)**3)))*(L/(rad**2))*dr
    #Add to lists
    mvals.append(m)
    rhovals.append(rhonew)
    Pvals.append(P)
    Lvals.append(L)
    Tvals.append(T)
```

# Problem 3

You probably noticed that the previous stellar structure solution did not give the correct temperature
and luminosity at the surface of the Sun. Let’s investigate a more complete consideration of energy
transport. Recall that the condition for convection is given by (see Lecture 9 notes):

$\frac{d log T}{d log P} \le \frac{3}{64 \pi \sigma G } \frac{P}{T^4} \frac{\kappa L}{M}$

What regions of the star are unstable to convection? (Hint: As in the previous problems, you can
approximate derivatives as differences.) (8 pts total)

We can break down the condition in terms of deltas:

$\Delta logT \le \frac{3}{64 \pi \sigma G } \frac{P}{T^4} \frac{\kappa L}{M} \Delta log P$

```python
dlogT = np.diff(np.log(Tvals))
dlogP = np.diff(np.log(Pvals))
is_stable = dlogT <=((3)/(64*pi*sigma*G)) * (np.array(Pvals[1:])/(np.array(Tvals[1:])**4)) * ((kappa*np.array(Lvals[1:]))/(np.array(mvals[1:]))) * dlogP
```

```python
changer = rvals[np.where(is_stable == True)[0][0]]
print(f'The sun is unstable to convection at the core and becomes stable at {changer:.2e} cm.')
```

```python
plt.rcParams['figure.figsize'] = (20, 10)
plt.plot(rvals, mvals, c =  'blue', label = 'Unstable', lw = 5, zorder = 10)
plt.plot(np.array(rvals[1:])[is_stable], np.array(mvals[1:])[is_stable], c = 'r', label = 'Stable', lw = 5, 
         zorder = 10)
plt.axvline(changer, zorder = 0,c = 'k', label = f'{changer:.2e} cm', lw = 5)
plt.xlabel('Radius [cm]')
plt.ylabel('Mass [g]')
plt.legend(fontsize = 'large')
plt.title('Mass vs Radius');
```

I will take this equation by equation.

**mass distribution**:

$\frac{dm}{dr} = 4 \pi r^2 \rho(r)$

We assume spherical symetry here, which breaks down in stars that have pulsations or regular changes in flux. Also any stars with any kind of turbulence in their atmospheres like cataclysmic variables.

We neglect the mass flow component of this equation and only use the static component. This neglects any time domain changes like acretion flows in contact binarie and winds in massive stars.

This neglect of mass flow assumes local thermodynamic equilibrium and requires that the mass velocity be zero.

**hydrostatic equilibrium**:

$\frac{dP}{dr} = \frac{-Gm(r) \rho(r)}{r^2}$

This equation assumes that gravity is totaly balanced by pressure. More specifically, it assumes that the acceleration due to gravity balances the pressure acceleration. Again, this is inherently an assumption that there is no time domain movement and the velocity can be approximated as zero, assuming local thermodynamic equilibrium. This neglects any small scale fluctuations due to pressure or gravity imbalances. Luckily, any divergence from hydrostatic equilibrium are restarted quickly in the types of stats we modeled, of course unless the white dwarf was part of a system that involved mass acretion or a supernovae.

Especially for pressure here, we are assuming that sudden changes are corrected. In stellar pulsations, variability is governed by sudden pressure changes that come from sudden opacity changes in a star. Pulsations like this would be neglected in this equation, and in our overall assumption of pressure balance.

**equation of state** 

dominated by ideal gas:

$P = \frac{\rho}{\mu m_H}kT$

For our ideal gas pressure equation, one of the most importnat assumptions we make is that particle collisions are perfectly inelastic. For main sequence stars, this assumption holds in most parts of most stars.

We also assume spherical symetry just like the rest of these equations. We assume that the quantities we expect to change across the radii of a stellar interior are on average spherically symetric.



equation of state for electron degeneracy with non-relativistic electrons:


$P = k_{NR}[\frac{\rho (r)}{\mu_R}]^{5/3}$

Here we assume the uncertainty principle holds and:

$\Delta x \Delta p \ge \frac{\hbar}{2}$

We assume that the pauli exclusion principle holds and:

$\bar{n}_i = \frac{1}{e^(\frac{\epsilon_i - \mu}{kT}) +1}$

where $\epsilon_i$ is the energy per particle and $\mu$ is the total chemical potential. For the non-relativistic case, we assume that:

$\epsilon_i = \frac{p_i}{2 m_e}$

We also assume that T does not change our equation of state, which holds for a white dwarf. 



**Energy transport**:

$\frac{dT}{dr} = \frac{-3 \kappa \rho}{64 \pi \sigma T^3} \frac{L}{r^2}$

We assumed here that energy transport was dominated by radiative diffusion. This assumes that radiation pressure comes from photons s.t.:

$P = \frac{4}{3} \frac{1}{c} \sigma_{sb} T^4$

We assume that this radiation pressure due to photons is the main source of outward pressure. Of course we know this is not always true. In problem one, we looked at a White Dwarf that does not have radiation pressure as a major driving force of pressure. 

This radiation pressure dominates main sequence stars, especially in their cores. This also depends on the ionization state of the star.

**Energy generation by thermal fusion**:

$\frac{dL}{dr} = 4 \pi r^2 \rho(r) [\epsilon_{nuc} - \epsilon{\nu}]$

We have to follow several assumptions for nuclear reactions to drive energy generation in stars:

1. energy and mass are NOT conserved

2. momentum IS conserved

3. charge IS conserved

4. number of nucleons IS conserved 

5. weak nuclear reactions conserve lepton number 

In this project, we assume that thermal fusion is the main energy generator, ignoring energy transport due to convection, although we touch on this in problem 3.
