Beyond the angular resolution uncertainty, what else might contribute uncertainty to the phys-
ical radius measurement? (1 pt)

$\rho(r) = [519(\frac{r}{R})^4 - 1630(\frac{r}{R})^3 + 1844 (\frac{r}{R})^2 -889 (\frac{r}{R}) + 155][\frac{g}{cm^3}]$ 

at r < 0.5R, and r $\approx$ 0.1 g/cm3 at r > 0.5R.

This fit comes mostly from astroseismology data,
which we will discuss later in the course.)

Here R = 6.96 x10^10 cm, the radius at the surface of the
Sun. 

Plot the enclosed mass profile m(r) of the Sun. Label the axes, and mark the average density
  and radius where (r) =  . (5 pts)

**answer** 

We are given the density profile and want the mass. Since

$\rho = \frac{m}{V}$

I will aproach this by breaking the sun up in to shells of some thickness dr. I can get the mass of a single one of these shells with:

$\rho(r) 4 \pi r^2 dr$

where $4 \pi r^2$ represents the surface area of a given shell with thickness dr of density $\rho$ at some inner stellar radiur r.

```python
def p(r):
    #radius in cm
    R = 6.96e10 
    if r<=(0.5*R):
        #Return the density in g/cm^3
        return ((519*((r/R)**4)) - (1630*((r/R)**3)) + (1844*((r/R)**2)) - (889*(r/R)) + 155)
    elif r>(0.5*R):
        #There is a set density 
        return 0.1
    #Adding some error statements
    elif r>(R):
        print('Please enter a value in cm less than the radius of the sun, 6.96x10^10 cm.')
    else:
        print('Please enter a positive value.')
```

```python
def shell_mass(r, dr):
    #return the mass of the shell
    return p(r)*4*np.pi*(r**2)*dr
```

```python
dr = 1e7
shells = np.arange(0,6.96e10, dr)
shell_masses = []
for r in shells:
    mass = shell_mass(r,dr)
    shell_masses.append(mass)
```

```python
plt.plot(shells, np.cumsum(shell_masses))
plt.xlabel('Radius[cm]')
plt.ylabel('Mass enclosed [g]')
plt.title('Mass Distribution');
```

```python
densities = []
for r in shells:
    densities.append(p(r))

plt.plot(shells, (densities), label = 'Density ')
pavg = np.mean(densities)
plt.axhline(pavg, c = 'k', label = 'Average Density', lw = 5, ls = '--')
plt.xlabel('Radius[cm]')
plt.ylabel('Density [g/cm^3]')
plt.title('Density Change')
plt.legend();
```

```python
#Get the radius where the acerage density occurs
ind = np.where(np.array(densities) == np.array(densities)[densities<pavg][0])[0][0]
ravg = shells[ind]

plt.plot(shells, np.cumsum(shell_masses))
plt.axvline(ravg, label = f'Radius of Average Density \nr = {ravg:.2e} cm \np = {round(pavg,2)} g/cm^3'
            , lw = 5, ls = '--', c = 'k')
plt.xlabel('Radius[cm]')
plt.ylabel('Mass enclosed [g]')
plt.title('Mass Distribution')
plt.legend(fontsize = 13);
```

# Problem 3

Assume a star made solely of ionized hydrogen

**a** Create (by sketch or computer) a plot of log density log() versus log temperature log(T)
(where log is the base-10 logarithm). 

Mark the boundaries between regions dominated by ideal
gas pressure, radiation pressure, and nonrelativistic electron degeneracy pressure (i.e., where
Pgas  =  Prad and Pgas = Pe. (5 pts total)

**answer**

From lecture 1/31:

For ideal gas,

ionizaed H, $\mu = 0.5$ 

$P = \frac{\rho}{\mu m_H}kT$

$P = \frac{\rho}{0.5 m_H}kT$

$P = \frac{2 \rho}{ m_H}kT$

For electron degeneracy pressure (nonrelativistic)

$P_e = k_{NR} (\frac{\rho}{\mu_e})^{2/3}$

fully ionized H means one proton, one electron. That's one electron per hydrogem molecule, so the ratio does not change and $\mu_e = \mu$ = 0.5

$P_e = k_{NR}({2 \rho})^{2/3}$

For radiation pressure,

$P_{rad} = \frac{4}{3c} \sigma_{sb}T^4$

Our equations of state are then:

$P_{rad} = \frac{4}{3c} \sigma_{sb}T^4$

$P_e = k_{NR}({2 \rho})^{2/3}$

$P_{gas} = \frac{2 \rho}{ m_H}kT$

Where the sun is governed by electron degeneracy, there is no temperature dependence. We scale with pressure as:

$P_e = k_{NR}({2 \rho})^{2/3}$

But we are interested in where this intersects with the ideal gas equation

$P_{gas} = \frac{2 \rho}{ m_H}kT$

or where 

$k_{NR}({2 \rho})^{2/3} = \frac{2 \rho}{ m_H}kT$


$T = \frac{k_{NR}}{k}\frac{m_H (2 \rho)^{2/3}}{2 \rho}$

**where $P_e = P_{gas}$**:

$T = \frac{k_{NR}}{k} m_H (2 \rho)^{2/3}$

Next, wwe will look at the intersection between the ideal gas and the radiation pressure.

**where $P_{gas} = P_{rad}$**

$\frac{2 \rho}{ m_H}kT = \frac{4}{3c} \sigma_{sb}T^4$

$T^3 = \frac{6 c k}{4 \sigma_{sb} m_H}\rho$

$T = [\frac{6 c k}{4 \sigma_{sb} m_H}\rho]^{1/3}$

```python
#Defining an array of values for density with correct units
rho = np.arange(0,40, .01)*(u.g/(u.cm**3))
logrho = np.log10(rho.value)
logrho
```

```python
#Getting T values for Pgas = Prad
Tpgprad = (((6*c.c*c.k_B)/(4*c.sigma_sb*(c.u)))*rho)**(1/3)
#Temparature in [K]
Tpgprad_K = Tpgprad.decompose()
Tpgprad_K
```

```python
#Getting T values in Pe = Pgas
knr = (c.h**2)/(20*c.m_e*(c.u**(5/3)))*((3/np.pi)**(2/3))
Tpepgas = (knr/(c.k_B))*(c.u)*((2*rho)**(2/3))
Tpepgas_K = Tpepgas.decompose()
Tpepgas_K
```

```python
plt.plot(logrho, np.log10(Tpgprad_K.value), label = r'$P_{gas} = P_{rad}$', lw = 5)
plt.plot(logrho, np.log10(Tpepgas_K.value), label = r'$P_{gas} = P_{e}$', lw = 5)

plt.grid()

plt.text(-1.5, 6, 'Ideal Gas Pressure', fontsize = 20)
plt.text(-.5, 4.5, 'Electron Degeneracy Pressure', fontsize = 20)
plt.text(-1.7, 7.9, 'Radiation Pressure', fontsize = 20)

plt.xlabel(r'log($\rho$)')
plt.ylabel('log(T)')
plt.legend(loc = 'upper right', fontsize = 20);
```
